```
train_n = data_t.iloc[0:20]
train_t = data_t.iloc[29:49]
train = pd.concat([train_n, train_t])
print(train)

validation_n = data_t.iloc[20:29]
validation_t = data_t.iloc[49:59]
validation = pd.concat([validation_n, validation_t])
print(validation)
```

```
validation.head()
```

```
print("Validation shape: ", validation.shape)
```

```
# initialize list of lists 
train_lbl = [['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'],
             ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T']] 

val_lbl = [['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'], ['N'],
           ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T'], ['T']]
# Create the pandas DataFrame 
train_lbl_df = pd.DataFrame(train_lbl, columns = [0]) 
val_lbl_df = pd.DataFrame(val_lbl, columns = [0]) 
```

```
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```
X_std = preprocessing.scale(train)
X_std_df = pd.DataFrame(X_std, columns=train.columns)
```

```
X_std_val = preprocessing.scale(validation)
X_std_val_df = pd.DataFrame(X_std_val, columns=validation.columns)
```

```
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
clf.fit(X_std_df, train_lbl_df)
```

```
predictions = clf.predict(X_std_val_df)
```

```
print(accuracy_score(val_lbl_df , predictions))
```

```
import matplotlib.pyplot as plt
```

```
min_estimators = 15
max_estimators = 175
X = X_std_df
y = train_lbl_df

error_rate=[]

randf = RandomForestClassifier(n_estimators=100, warm_start=True, max_depth=2, oob_score=True, random_state=0)

for i in range(min_estimators, max_estimators + 1):
  randf.set_params(n_estimators=i)
  randf.fit(X, y)
  # Record the OOB error for each `n_estimators=i` setting.
  oob_error = 1 - randf.oob_score_
  error_rate.append((i, oob_error))
```

```
xs = []
ys = []
for clf_err in error_rate:
  xs.append(clf_err[0])
  ys.append(clf_err[1])

plt.plot(xs, ys)
plt.xlim(min_estimators, max_estimators)
plt.xlabel("n_estimators")
plt.ylabel("OOB error rate")
plt.show()
```
